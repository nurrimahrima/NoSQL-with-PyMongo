### 1. Make new collection, **clean_movies** that is similar with **movies** collection in **sample_mflix** database using **movies_initial** collection entities (projecting_queries pymongo). 
***

First, we need to define different fields from **movies_initial** to become equal with fields in **movies**, such as:
1. imdb
2. released
3. tomatoes
4. awards
5. runtime

```python
#show all unique keys in every documents in 'movies_initial' collection 
x = []
for doc in col.find({}):
    x.append(doc.keys())
x[:5]
```

For reasons of there might be big possibility of different length of keys or any typos of keys in every document, we add some 'condition' in every fields where the output will be the equal value ('eq') with **movies** or it will not show up at all ("nonExistinField") adjusting with all fields in each document in **movies**.

```python
#define imdb
imdb = list(col1.find({},{'_id':0, 'imdb':1}))
imdb_list = []
print(len(imdb))
for pin in imdb:
  imdb_list.append(list(pin.values())[0])
  
imdb_list[:5]
```

```python
imdb=[]
for doc in imdb_list:
  imdb.append(doc["id"])
imdb[:5]
```

```python
#define awards, runtime, released, tomatoes in pipeline directly
#Projecting to run new collection
pipeline = [
    {
        '$project' : {
            ### this code is to include the keys for a new collection on database ###
            ### 1 for include , 0 for exclude the keys ###
            ### give some condition between including when exist and excluding when nonexist
            'plot' : {"$cond": [{"$eq": ["$plot", ""]}, "$nonExistinField", "$plot"]},
            'genres' : {"$cond": [{"$eq": ["$genre", ""]}, "$nonExistinField", {'$split' : ['$genre', ', ']}]},
            'runtime': {"$cond": [{'$eq': ["$runtime", ""]}, 0 ,{'$arrayElemAt':[{'$split': [ "$runtime", " min" ]}, 0]}]},
            'cast' : {"$cond": [{"$eq": ["$cast", ""]}, "$nonExistinField", {'$split' : ['$cast', ', ']}]},
            'num_mflix_comments' : {"$cond": [{"$eq": [{'$sum':1}, ""]}, "$nonExistinField", {'$sum':1}]},
            'poster' : {"$cond": [{"$eq": ["$poster", ""]}, "$nonExistinField", "$poster"]},
            'title' : {"$cond": [{"$eq": ["$title", ""]}, "$nonExistinField", "$title"]},
            ### renaming the keys ###
            'fullplot' : {"$cond": [{"$eq": ["$fullplot", ""]}, "$nonExistinField", '$fullplot']},
            ### here we split some keys into an array value, renaming as well ###
            'languages' : {"$cond": [{"$eq": ["$language", ""]}, "$nonExistinField", {'$split' : ['$language', ', ']}]},
            'countries' : {"$cond": [{"$eq": ["$country", ""]}, "$nonExistinField", {'$split' : ['$country', ', ']}]},
            'released': {"$cond": [{"$eq": ["$released", ""]}, "$nonExistinField", 
                                   {'$cond': {
                    'if': {'$ne': ["$released", ""]},
                    'then': {
                        '$dateFromString': {
                            'dateString': "$released"
                        }
                    },
                    'else': ""}}]},
            'directors' : {"$cond": [{"$eq": ["$director", ""]}, "$nonExistinField", {'$split' : ['$director', ', ']}]},
            'writers' : {"$cond": [{"$eq": ["$writer", ""]}, "$nonExistinField", {'$split' : ['$writer', ', ']}]},
            'rated' : {"$cond": [{"$eq": ["$rating", ""]}, "$nonExistinField", '$rating']},
            'awards' : {'nominations': {'$arrayElemAt':[{'$split':[{'$arrayElemAt':[{'$split': [ "$awards", " nomination" ]}, -2]},' ']},-1]},
                       'text': "$awards",
                       'wins': {'$arrayElemAt':[{'$split':[{'$arrayElemAt':[{'$split': [ "$awards", " win" ]}, -2]},' ']},-1]}
                      },
            'lastupdated' : {"$cond": [{"$eq": ["$lastupdated", ""]}, "$nonExistinField", '$lastupdated']},
            'year' : {"$cond": [{"$eq": ["$year", ""]}, "$nonExistinField", "$year"]},
            ### adding embeded documents ###
            ### grouping the several keys into embeded documents by one keys ###
            'imdb' : {
                'id' : '$imdbID',
                'rating' : '$imdbRating',
                'votes' : '$imdbVotes',
                },
            'type' : {"$cond": [{"$eq": ["$type", ""]}, "$nonExistinField", "$type"]},
            'metacritic' : {"$cond": [{"$eq": ["$metacritic", ""]}, "$nonExistinField", "$metacritic"]},
            'tomatoes' : {
                'lastUpdated': '$lastupdated',
                'viewer': {'meter':'', 'numReviews':'', 'rating':''},
                'dvd':'',
                'fresh':'',
                'rotten':'',
                'production':''
            }
        }
        },
        {
            '$match':{'imdb':{"$in":[dict(sorted(imdb_list[x].items(), key=lambda x: x[0])) for x in range(len(imdb_list))]}}
        },
    
    {
        #naming new collection as output data
        '$out' : 'clean_movies_nurrimah'
    }
    
]

clear_output ()
pprint.pprint(list(client1.sample_mflix.movies_initial.aggregate(pipeline)))
```

In this code below, we will check the equality of the fields of two collection with some anagram function.

```python
#define the two words, where the first is fields in 'clean_movies', and the second is fields in 'movies'
def isAnagram(s1, s2):
  sortedWord1 = sorted(s1)
  sortedWord2 = sorted(s2)
  if sortedWord1 == sortedWord2:
    print('Both fields are the same')
  else:
    print("Both fields are not the same")

word1 = clean
word2 = mov

isAnagram(word1, word2)
```
